### 팔레트 종류 
* deep: 진한 색상의 팔레트* 
muted: 부드러운 색상의 팔레
* pastel: 연한 색상의 팔레트
* bright: 밝은 색상의 팔레트
* dark: 어두운 색상의 팔레트
* colorblind: 색맹을 고려한 팔레트
* Set1, Set2, Set3: 범주형 데이터에 적합한 팔레트
* Paired: 순차적인 색상 팔레트
* Accent: 강조를 위한 팔레트
* Inferno, Magma, Plasma, Viridis: 연속형 데이터에 적합한 팔레트
* cubehelix: 연속적인 그레이스케일 팔레트

### 팔레트 특징과 용도 
* deep: 진한 색상으로 구성되어 있어 대비가 강한 편입니다. 범주형 데이터 시각화에 적합합니다.
* muted: 부드러운 색상으로 구성되어 있어 시각적으로 편안합니다. 범주형 데이터 시각화에 적합합니다.
* pastel: 연한 색상으로 구성되어 있어 부드러운 느낌을 줍니다. 범주형 데이터 시각화에 적합합니다.
* bright: 밝은 색상으로 구성되어 있어 시각적으로 강렬합니다. 범주형 데이터 시각화에 적합합니다.
* dark: 어두운 색상으로 구성되어 있어 고급스러운 느낌을 줍니다. 범주형 데이터 시각화에 적합합니다.
* colorblind: 색맹을 고려하여 설계된 팔레트로, 모든 사람이 구분할 수 있는 색상으로 구성되어 있습니다.있는 색상으로 구성되어 있습니다.

```python
df_sort = df.sort_values(by=["운동을 할 만한 장소가 없어서"], axis=0, ascending=False) # 내림차순 정렬
df_sort
```

```python
df_sort.rename(columns={
    df_sort.columns[0] : "분류",
    df_sort.columns[1] : "장소부족",
    df_sort.columns[2] : "비용부족" }, inplace=True )
df_sort
```

#### MinMaxScaler(표준화)
* 모든 특성이 정확하게 0과 1사이에 위치하도록 데이터를 변경한다
* 가장 작은값을 0, 가장 큰값을 1로 두고 결과를 뽑아준다

```python
df_sort.describe()
# 각각의 min, max 값이 다르므로 두 값을 평준화시켜야 함!
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
movie = {"daum" : [2, 4, 6, 8, 10], "naver" : [1, 2, 3, 4, 5] }
mv = pd.DataFrame(movie)
mv
```

```python
min_max_scaler = MinMaxScaler()
mv_sc = min_max_scaler.fit_transform(mv) # 척도를 같게 만들어줌
mv_sc
```

```python
pd.DataFrame(mv_sc, columns=["daum", "naver"])
```

```python
# 척도 같게 만들어주기 
min_max_s = MinMaxScaler()
min_max_df = min_max_s.fit_transform(df_sort[["장소부족", "비용부족"]])
min_max_df
```

```python
# 데이터 프레임으로 변경하기 
df_replace = pd.DataFrame(min_max_df, columns=["장소부족", "비용부족"])
df_replace
```

```python
df_replace.describe()
# min, max의 값이 평준화 된 것을 확인할 수 있음 
```

```python
df_replace["분류"] = df_sort["분류"].values
df_replace
```

```python
df_sort = df_replace[["분류", "장소부족", "비용부족"]]
df_sort
```

```python
# 가중치 부여하기 
df_sort["종합점수"] = df_sort["장소부족"]*0.6 + df_sort["비용부족"]*0.4
df_top5 = df_sort.sort_values(by="종합점수", ascending=False).head() # 종합점수를 기준으로 정렬
df_top5
```

```python
# 시각화 하기 
g = sns.barplot(x="분류", y="종합점수", data=df_top5, hue="분류")
```

```python
df["대분류"].isin(['성별'])
```

```python
df[df["대분류"].isin(['성별'])]
```

```python
df_gender = df[df["대분류"].isin(['성별'])]
df_gender
```

```python
features = [
    '운동을 할 충분한 시간이 없어서', '함께 운동을 할 사람이 없어서'
]
```

```python
pg = sns.PairGrid(df_gender, y_vars="분류", x_vars=features, height=5, hue="분류")
pg.map(sns.barplot)
```

```python
df_age = df[df["대분류"].isin(["연령별"]) ]
df_age
```

```python
pg = sns.PairGrid(df_age, y_vars="분류", x_vars=features, height=5, hue="분류")
pg.map(sns.barplot)
```

```python
df_age["운동할 수 있음"] = df_age["운동을 할 충분한 시간이 없어서"].apply(lambda x: 100-x)
df_age["함께 운동할 수 있음"] = df_age["함께 운동을 할 사람이 없어서"]
f = ["운동할 수 있음", "함께 운동할 수 있음"]
pg = sns.PairGrid(df_age, y_vars="분류", x_vars=f, height=5, hue="분류")
pg.map(sns.barplot)
```

```python
df_age["종합점수"] = df_age["운동할 수 있음"] + df_age["함께 운동할 수 있음"]
f = ["운동할 수 있음", "함께 운동할 수 있음", "종합점수"]
pg = sns.PairGrid(df_age, y_vars="분류", x_vars=f, height=5, hue="분류")
pg.map(sns.barplot)
df_age[["분류", "종합점수"]]
```
