```python
import pandas as pd
df = pd.read_csv('../data_set/2.데이터 클린징/loan_train_quiz_cleaning.csv')
df.head()
```

### 변수
* label : 정답, 종속변수
* feature : 속성, 독립변수 

```python
df.columns
```

```python
features = ['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']
label = 'Loan_Status'
X, y = df[features], df[label]
```

```python
y
```

```python
X
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split 
#주어진 데이터 세트를 훈련 데이터 세트(train set)와 테스트 데이터 세트(test set)로 나누는 기능
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, len(y_train), len(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier # 의사 결정 트리(Decision Tree) 모델을 생성하고 훈련하는 과정
dtc = DecisionTreeClassifier() #DecisionTreeClassifier 클래스의 인스턴스를 생성하여 dtc 변수에 할당
dtc.fit(X_train, y_train) 
# 생성된 Decision Tree 분류기 모델에 X_train 데이터를 입력으로, y_train 데이터를 타깃으로 하여 학습시킨다.
```

```python
dtc.predict(X_test) # X_test 데이터에 대한 예측을 수행
```

```python
pred = dtc.predict(X_test)
```

```python
y_test
```

```python
dtc.score(X_test, y_test)
# 모델 객체의 내장 메서드를 사용하여 모델의 성능을 평가
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(pred, y_test) 
# accuracy_score 함수를 사용하여 모델의 예측값과 실제 타겟값을 비교하여 정확도를 계산하는 부분
```
